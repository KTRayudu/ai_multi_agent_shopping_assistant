````python
import os
import pandas as pd
import pyarrow.dataset as ds
import json

# Create the data folder
os.makedirs("data", exist_ok=True)

print("=" * 60)
print("STEP 1: Loading Electronics Metadata (Parquet)...")
print("=" * 60)

# Read all metadata shards (full-00000-of-00010.parquet, etc.)
meta_dataset = ds.dataset(
    "hf://datasets/McAuley-Lab/Amazon-Reviews-2023/raw_meta_Electronics",
    format="parquet"
)
meta_table = meta_dataset.to_table()
meta_df = meta_table.to_pandas()
print(f"✅ Loaded {len(meta_df):,} metadata items.")

# Save the FULL metadata as JSONL (the notebook expects this file)
meta_df.to_json("data/meta_Electronics.jsonl", orient="records", lines=True)
print("✅ Saved FULL metadata to 'data/meta_Electronics.jsonl'")

print("\n" + "=" * 60)
print("STEP 2: Filter & Sample 1000 Items (NO need to download 20GB)...")
print("=" * 60)

# --- Filter exactly as the bootcamp notebook does ---
meta_df = meta_df[meta_df['rating_number'] > 100]      # Popular items
meta_df = meta_df[meta_df['price'].notna()]            # Items with price

# Create the 'description' column (title + features)
def create_description(row):
    features = ' '.join(row['features']) if isinstance(row['features'], list) else ''
    return f"{row['title']} {features}"

meta_df['description'] = meta_df.apply(create_description, axis=1)

# Extract first image URL
def get_first_image(row):
    imgs = row.get('images', [])
    if imgs and isinstance(imgs, list):
        img = imgs[0]
        return img.get('large') or img.get('hi_res') or ''
    return ''

meta_df['image'] = meta_df.apply(get_first_image, axis=1)

# Sample 1000 items
sample_df = meta_df.sample(n=1000, random_state=20)

# Save the SAMPLE (this is what the notebook actually uses)
sample_df.to_json(
    "data/Electornics_2022_2023_with_category_ratings_100_sample_1000.jsonl",
    orient="records",
    lines=True
)
print(f"✅ Saved {len(sample_df)} sampled items to 'data/Electornics_2022_2023_with_category_ratings_100_sample_1000.jsonl'")

print("\n" + "=" * 60)
print("STEP 3: (Optional) Loading ONLY reviews for the 1000 sampled items...")
print("=" * 60)

# Get the parent_asins of our sample
sample_asins = set(sample_df['parent_asin'])

# Read reviews but FILTER at the Parquet level – only downloads matching rows!
review_dataset = ds.dataset(
    "hf://datasets/McAuley-Lab/Amazon-Reviews-2023/raw_review_Electronics",
    format="parquet"
)

# Apply filter before loading (massively reduces memory/time)
filtered_reviews = review_dataset.to_table(
    filter=ds.field('parent_asin').isin(sample_asins)
)
review_df = filtered_reviews.to_pandas()

# Save reviews sample
review_df.to_json("data/Electronics_reviews_sample.jsonl", orient="records", lines=True)
print(f"✅ Saved {len(review_df)} matching reviews to 'data/Electronics_reviews_sample.jsonl'")

print("\n" + "=" * 60)
print("🎉 DONE! All required files are in the 'data/' folder.")
print("=" * 60)
````

````python
import os
import json
from datasets import load_dataset

# Load your sample ASINs
sample_df = pd.read_json("data/Electornics_2022_2023_with_category_ratings_100_sample_1000.jsonl", lines=True)
sample_asins = set(sample_df['parent_asin'])

print("Loading reviews in streaming mode (filtering as we go)...")
review_ds = load_dataset(
    "McAuley-Lab/Amazon-Reviews-2023",
    "raw_review_Electronics",
    split="full",
    streaming=True          # Does NOT download the whole dataset
)

# Write matching reviews to file
with open("data/Electronics_reviews_sample.jsonl", "w") as f:
    count = 0
    for record in review_ds:
        if record['parent_asin'] in sample_asins:
            f.write(json.dumps(record) + "\n")
            count += 1
            if count % 100 == 0:
                print(f"  Collected {count} reviews...")
print(f"✅ Saved {count} reviews to 'data/Electronics_reviews_sample.jsonl'")
````

## Filter Items that have been observed for the first time in year 2022 or later.

````python
def filter_data(data: dict) -> dict:
    filter = False
    if int(data['details']['Date First Available'][-4:]) < 2022:
        filter = True

    return filter
````

````python
with open("../../data/meta_Electronics.jsonl", 'r') as fp:
    with open("../../data/meta_Electronics_2022_2023.jsonl", 'a', encoding='utf-8') as fp_out:
        with open("../../data/meta_Electronics_2022_2023_no_date.jsonl", 'a', encoding='utf-8') as fp_out_no_date:
            i = 0
            for line in fp:
                data = json.loads(line.strip())
                try:
                    filter = filter_data(data)
                    if not filter:
                        json.dump(data, fp_out)
                        fp_out.write('\n')
                        fp_out.flush()
                except:
                    json.dump(data, fp_out_no_date)
                    fp_out_no_date.write('\n')
                    fp_out_no_date.flush()
                i += 1
                if i % 10000 == 0:
                    print(f"Processed {i} lines")
````

## Split the items into two categories: "has main category", "does not have main category"

````python
def filter_category(data: dict) -> dict:
    filter = False
    if data['main_category'] == None:
        filter = True

    return filter
````

````python
with open("../../data/meta_Electronics_2022_2023.jsonl", 'r') as fp:
    with open("../../data/meta_Electronics_2022_2023_with_category.jsonl", 'a', encoding='utf-8') as fp_out:
        with open("../../data/meta_Electronics_2022_2023_no_category.jsonl", 'a', encoding='utf-8') as fp_out_no_category:
            for line in fp:
                data = json.loads(line.strip())
                if not filter_category(data):
                    json.dump(data, fp_out)
                    fp_out.write('\n')
                    fp_out.flush()
                else:
                    json.dump(data, fp_out_no_category)
                    fp_out_no_category.write('\n')
                    fp_out_no_category.flush()
````
